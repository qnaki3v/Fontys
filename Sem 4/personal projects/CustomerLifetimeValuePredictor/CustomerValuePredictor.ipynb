```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1 = pd.read_csv('retail1.csv',encoding='latin1')
df_2 = pd.read_csv('retail2.csv')
```

```python
df_1.info()
df_1.head()
```

```python
df_2.info()
df_2.head()
```

```python
df_1.rename(columns={'InvoiceNo': 'Invoice', 'UnitPrice': 'Price'}, inplace=True)
df_1.head()
```

```python
df_2.rename(columns={'Customer ID': 'CustomerID'}, inplace=True)
df_2.head()
```

```python
df_1['InvoiceDate'] = pd.to_datetime(df_1['InvoiceDate'])  # Convert to datetime
df_1['InvoiceDate'] = df_1['InvoiceDate'].dt.strftime('%d/%m/%Y %H:%M:%S')
print(df_1['InvoiceDate'].head())  # Display the original and formatted columns
df_2['InvoiceDate'] = pd.to_datetime(df_2['InvoiceDate'])  # Convert to datetime
df_2['InvoiceDate'] = df_2['InvoiceDate'].dt.strftime('%d/%m/%Y %H:%M:%S')
print(df_2['InvoiceDate'].head())  # Display the original and formatted columns
```

```python
unique_count = df_1['CustomerID'].nunique()
print(f'Number of unique products: {unique_count}')
print(df_1['CustomerID'].isna().sum())  # Shows how many NaN values are present

unique_count = df_2['CustomerID'].nunique()
print(f'Number of unique products: {unique_count}')
print(df_2['CustomerID'].isna().sum())  # Shows how many NaN values are present
```

```python
df_1['CustomerID'] = df_1['CustomerID'].apply(lambda x: str(int(x)) if pd.notna(x) else 'Unknown')
df_1['CustomerID'].info
unique_count = df_1['CustomerID'].nunique()
print(f'Number of unique products: {unique_count}')
print(df_1['CustomerID'].isna().sum())  # Shows how many NaN values are present

df_2['CustomerID'] = df_2['CustomerID'].apply(lambda x: str(int(x)) if pd.notna(x) else 'Unknown')
df_2['CustomerID'].info
unique_count = df_2['CustomerID'].nunique()
print(f'Number of unique products: {unique_count}')
print(df_1['CustomerID'].isna().sum())  # Shows how many NaN values are present
# there is 1 more record in each dataset because NAN previously was not an unique input but after it was
# convert to "Unknown" it is.
```

```python
combined_df = pd.concat([df_1, df_2], ignore_index=True)

# Check the result
print(combined_df.head())
print(combined_df.shape)
```

```python
product_sales = combined_df.groupby('Description')['Quantity'].sum().reset_index()

# Sort by total quantity sold in descending order
product_sales = product_sales.sort_values(by='Quantity', ascending=False)

# Select top 10 products
top_10_products = product_sales.head(10)

# Create the horizontal bar chart
plt.figure(figsize=(19, 6))
plt.barh(top_10_products['Description'], top_10_products['Quantity'], color='#dda15e')

plt.xlabel('Quantity Sold')
plt.ylabel('Product Description')
plt.title('Top 10 Most Sold Products')

plt.gca().invert_yaxis()  # Invert y-axis

# Add the quantity values to the bars
for index, value in enumerate(top_10_products['Quantity']):
    plt.text(value, index, f'{value:,.0f}', va='center')
```
